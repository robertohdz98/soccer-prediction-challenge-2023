# Compare training set and prediction set

```python
test_set = pd.read_excel("datasets/raw/PredictionSet_2023_01_31.xlsx")
test_set
```

```python
teams_test_set = set(test_set["HT"].to_list() + test_set["AT"].tolist())
print("Number of teams in test set:", len(teams_test_set))
teams_test_set
```

```python
training_set = pd.read_parquet("datasets/processed/training_set_processed.parquet")
teams_training_set = set(training_set["HT"].to_list() + training_set["AT"].tolist())

print("Number of teams in training set:", len(teams_training_set))
teams_training_set
```

## Which teams in test set are not present in training set???

```python
n = 0
for t in teams_test_set:
    if (t in training_set["HT"].values) or (t in training_set["AT"].values):
        n = n+1
    else:
        print(t)
n
```
